##### **Problem 05:**
Write a function that takes a number and returns a vector like the previous problem. For example, if the input is 5, then the output should be `[25, 16, 9, 4, 1]`. If the input is 10, then the output should be `[100, 81, 64, 49, 36, 25, 16, 9, 4, 1]`.

##### **Problem 08:**
Plot $x + y + z = 1$ for $x, y \in [-3, 3]$ with interval 0.5. Use meshgrid to generate the grid. You can use any built-in function to plot the surface.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x = np.arange(-3, 3.5, 0.5)
y = np.arange(-3, 3.5, 0.5)
X, Y = np.meshgrid(x, y)

Z = 1 - X - Y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z,cmap='viridis', edgecolor='none')


ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Plot of x + y + z = 1')

plt.show()
```

##### **Problem 10:**
Compute 2nd order Taylor approximation of $e^{2x}$ and $e^{3x}$, and $e^{5x}$ around $a = 0$ for $x \in [-1, 1]$ with interval 0.01. Then, determine which of the following is more accurate:
> 1. Calculating the Taylor approximation separately for each $e^{2x}$, $e^{3x}$ then multiplying them together.
> 2. Calculating the Taylor approximation for $e^{5x}$.

Use mean square error (MSE) to determine the accuracy. Also, plot the Taylor approximation for 1st and 2nd method and the original function in the same plot for visual comparison.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.arange(-1, 1.01, 0.01)

# Define the original functions
def f1(x):
    return np.exp(2*x)

def f2(x):
    return np.exp(3*x)

def f3(x):
    return np.exp(5*x)

# Define the Taylor approximations
def taylor_approx_f1(x):
    return 1 + 2*x + (4*(x**2))/2  # 2nd order approximation of e^(2x)

def taylor_approx_f2(x):
    return 1 + 3*x + (9*(x**2))/2  # 2nd order approximation of e^(3x)

def taylor_approx_f3(x):
    return 1 + 5*x + (25*(x**2))/2 # 2nd order approximation of e^(5x)

# Calculate the Taylor approximations for each method
method1_approx = taylor_approx_f1(x) * taylor_approx_f2(x)  # Multiplying separate approximations
method2_approx = taylor_approx_f3(x)  # Direct approximation of e^(5x)

# Calculate the MSE for each method
mse_method1 = np.mean((f3(x) - method1_approx)**2)
mse_method2 = np.mean((f3(x) - method2_approx)**2)

# Print the MSE values
print("MSE for Method 1:", mse_method1)
print("MSE for Method 2:", mse_method2)

# Plot the results
plt.plot(x, f3(x), label='Original Function (e^(5x))')
plt.plot(x, method1_approx, label='Method 1 (e^(2x) * e^(3x))')
plt.plot(x, method2_approx, label='Method 2 (e^(5x))')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Taylor Approximations of e^(5x)')
plt.legend()
plt.grid(True)
plt.show()
```
